```python
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)

def flash(content):
    completion = client.chat.completions.create(
        model="google/gemini-2.0-flash-thinking-exp:free",
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": content},
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": "https://st3.depositphotos.com/13404340/18014/i/450/depositphotos_180140528-stock-photo-collection-of-objects-in-chaos.jpg"
                        },
                    },
                ],
            }
        ],
    )
    # print(completion.choices[0].message.content)
    if completion.choices and completion.choices[0].message:
        output = completion.choices[0].message.content
        # print(output)
        # print(type(content))
    else:
        output = f"Solution: {content}\nCritique:"
        print("No message content found.")

    return output

# print(flash("Solve the problem in the image."))

def meta(content):
    completion = client.chat.completions.create(
        model="meta-llama/llama-3.2-90b-vision-instruct:free",
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": content},
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": "https://st3.depositphotos.com/13404340/18014/i/450/depositphotos_180140528-stock-photo-collection-of-objects-in-chaos.jpg"
                        },
                    },
                ],
            }
        ],
    )
    # print(completion.choices[0].message.content)
    if completion.choices and completion.choices[0].message:
        output = completion.choices[0].message.content
        # print(output)
        # print(type(content))
    else:
        output = f"Solution: {content}\nCritique:"
        print("No message content found.")

    return output
```

```python
def write_to_file(input_string):
    # Open a text file in write mode
    with open("output.txt", "a") as file:
        # Write the input string to the file
        file.write(input_string + "\n")
        # Append "================" at the end
        file.write(
            "=================================================================================\n"
        )
```

```python
def iterative_refinement(problem, max_iterations=5):
    iteration = 0
    solution = problem

    while iteration < max_iterations:
        print(f"\n--- Iteration {iteration + 1} ---")
        write_to_file(f"\n--- Iteration {iteration + 1} ---")
        # LLM-A generates a response
        solution = flash(
            f"Task: {problem}\nSolution:"
        )
        write_to_file(solution)
        # print(f"Gemini Flash's Response:\n{solution}")

        # # LLM-B critiques the response
        critique = meta(
            f"Solution: {solution}\nCritique:"
        )
        # # print(f"Meta Critique:\n{critique}")
        write_to_file(critique)

        # # Evaluate stopping condition (can refine further logic here)
        if "no issues" in critique.lower():
            print("Consensus reached.")
            break

        # # Prepare input for next iteration
        problem = critique  # Alternatively, merge critique with original task
        iteration += 1

    return solution


# Example usage
problem = "How many objects are in image?"
final_solution = iterative_refinement(problem)
print(f"\nFinal Solution:\n{final_solution}")
```
