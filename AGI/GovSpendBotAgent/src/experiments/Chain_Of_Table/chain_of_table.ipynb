```python
import pandas as pd
from llama_index.llms.openrouter import OpenRouter
from llama_index.core.llms import ChatMessage

# Initialize OpenRouter (as you already have)
llm = OpenRouter(
    api_key="",
    max_tokens=256,
    context_window=4096,
    model="meta-llama/llama-3.1-70b-instruct:free",
    # model="undi95/toppy-m-7b:free",
)

# Read the CSV file using pandas (or use other methods to extract data)
# df = pd.read_csv("/Users/manojkl/Downloads/expense_data_1.csv")

# Convert the DataFrame into a text format that can be used in prompts
csv_text = df.to_string(index=False)  # Converts to string without row indices

# Prepare a message to pass to the model (assuming you are using a prompt-based approach)
user_message = ChatMessage(
    role="user",
    content=f"Here is the data from my CSV:\n{csv_text}\n what is the total food amount on date 3/1/2022? Don't give me program instead analyze the data and give me result, also give python program to plot food cost over the time",
)

# Use the LLM to process the message
response = llm.chat([user_message])

# Print the model's response
print(response)
```
